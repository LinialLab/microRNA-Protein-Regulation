```python
import pandas as pd
from sklearn.metrics import classification_report 
```

#### Our model predictions
* Includes zebrafish and thale
* `bool_mir_target_predicted`

```python
df_candidates = pd.read_csv("export-preds_30_thale_zebra_candidate_mirgene.csv")

df_candidates
```

```python
# df_ts_fish = pd.read_csv("targetScan_Summary_Counts/fish_Summary_Counts.tsv",sep="\t") # ORIG
df_ts_fish = pd.read_csv("targetScanSummaryCounts/zfish_Summary_Counts.txt",sep="\t")
print(df_ts_fish.nunique())
df_ts_fish
```

```python
## lowercasing removes some cases. str strip - no affect
fish_ts_gene_targets = df_ts_fish["Gene Symbol"].str.lower().str.strip().unique()
print(len(fish_ts_gene_targets))
```

```python
# df_candidates.columns

df_candidates.groupby(["species"])[['bool_mir_target', 'bool_mir_target_predicted']].agg(["mean","sum","size"])
```

* Note: some gene names missing (""), and some will lack match
* We will evaluate only on matches genes

Warning - there are multiple proteins per gene!

```python
df_candidates.sort_values('probability_1',ascending=False,inplace=True)

df_candidates.loc[df_candidates["bool_mir_target"]==0].head(10).iloc[:,0:8]
```

```python
df_fish = df_candidates.loc[df_candidates["species"]=="zebrafish"][['Gene names  (primary )','Protein names', 'bool_mir_target', 'bool_mir_target_predicted',
       'probability_1']].drop_duplicates()
## will drop many many rows#.drop_duplicates(subset=['Gene names  (primary )'])
###
## from 2
df_fish = df_fish.loc[df_fish['Gene names  (primary )']!=""].dropna() ## dropna removes 2,500 rows!
df_fish.sort_values('probability_1',ascending=False,inplace=True)

df_fish
```

```python
df_fish["targetscan_pred"] = (df_fish['Gene names  (primary )'].str.lower().str.strip().isin(fish_ts_gene_targets)).astype(int)
df_fish["targetscan_pred"].agg(["sum","mean","count"]).round(2)
```

### Candidates

```python
df_candidates.loc[df_candidates["bool_mir_target"]==0].head(10).iloc[:,0:8]
```

#### Agreement between model predictions and targetscan predictions
* Note that we're not taking gene vs protein into account - could improve, e.g. by taking max per gene

```python
df_fish[["targetscan_pred","bool_mir_target_predicted","bool_mir_target"]].corr().round(3)
```

Correlation without low confidence targets - against conserved families
* https://www.targetscan.org/cgi-bin/targetscan/data_download.cgi?db=fish_62
* Conserved Family Info - positions of conserved and nonconserved sites corresponding to **conserved miRNA families**


```python
df_fish_confident.columns
```

```python
df_fish_confident = pd.read_csv("targetScanSummaryCounts/fish_Conserved_Family_Info.txt.zip",sep="\t")
print(df_fish_confident.shape[0])
df_fish_confident["Gene Symbol"] = df_fish_confident["Gene Symbol"].str.lower().str.strip()
df_fish_confident = df_fish_confident[['miR Family', "Gene Symbol", 'Species ID',
       'UTR start', 'UTR end', 'MSA start', 'MSA end', 'Seed match']].drop_duplicates()
print(df_fish_confident.shape[0], "drop dupes", "genes:",df_fish_confident["Gene Symbol"].nunique() )

# ## filter by cases with at leat X miRNA targetting? This isn't the same as multiple experiments.. 
# ## Doesn't improve confidence (when filtering like this at least)
# df_fish_confident["count"] = df_fish_confident.groupby(["Gene Symbol"])['miR Family'].transform("count")
# df_fish_confident = df_fish_confident.loc[df_fish_confident["count"]>2]
# print(df_fish_confident.shape[0], "drop by count", "genes:",df_fish_confident["Gene Symbol"].nunique() )

fish_ts_gene_targets_confident = df_fish_confident["Gene Symbol"].str.lower().str.strip().unique()
print(len(fish_ts_gene_targets_confident),"# confident genes")
df_fish["targetscan_pred_confident"] = (df_fish['Gene names  (primary )'].str.lower().str.strip().isin(fish_ts_gene_targets_confident)).astype(int)
df_fish.mean()
```

```python
## check without low confidence targets
df_fish[["targetscan_pred","targetscan_pred_confident","bool_mir_target_predicted","bool_mir_target"]].corr().round(3)
```

```python
print(classification_report(y_true=df_fish["targetscan_pred"],
    y_pred=df_fish["bool_mir_target_predicted"]))
```

##### We see that the predictions have much better "agreement" with targetscan, than the raw experimental labels, mainly due to much higher recall!

```python
print(classification_report(y_true=df_fish["targetscan_pred"],
    y_pred=df_fish["bool_mir_target"]))
```
