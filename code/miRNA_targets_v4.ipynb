### drop leaks
* proteins without confirmed/manual/experimental existence, won';t have experiemntal validation of miRNA binding!

````python
print(df.shape[0])
df = df.loc[df['Protein existence'].str.contains("evidence",case=False,na=False)]
print(df.shape[0])
##Would just drop all the -1s .. # df = df.dropna(subset=["Target Gene"])
# print(df.shape[0])
````

#### additional data munge
* change ";" to make for easier tokenizing

````python
df = df.replace(";",". ",regex=True)
df = df.loc[df['Gene names']!=""]
DROP_COLS = ["Target Gene (Entrez ID)","Target Gene (Entrez Gene ID)","Target Gene","Organism","Status"] +\
['Entry', 'Entry name','Gene names', 'Gene names (Primary)',"Protein existence"] 
df.drop(columns=DROP_COLS,errors="ignore",inplace=True)
df.drop_duplicates(inplace=True)
df["total_nan"] = df.isna().sum(axis=1)
````

````python
df["bool_mir_target"] = (df["mir_target"]> -1).astype(int)
````
