* Get "higher quality" gene/protein targets

````python
import pandas as pd
````

````python
EVIDENCE_COLS = ['miRTarBase ID','Experiments','References (PMID)']
````

````python
df = pd.read_csv("hsa_MirTar_human.csv.gz")
df["mir_target"] = 0 ## default, will be updated to 1 for "high confidence" targets
print(df.nunique())
df
````

````python
df["Support Type"].value_counts(normalize=True) ### 97% weak
````

Get # unique times an interaction/target gene was validated (regardless of type of validation)

````python
evidence_cases = df.groupby("Target Gene (Entrez Gene ID)")[EVIDENCE_COLS].nunique().add_prefix("gene_nuniq_")
evidence_cases
````

````python
df = df.merge(evidence_cases,on=["Target Gene (Entrez Gene ID)"],how="left")
df
````

````python
df[["gene_nuniq_References (PMID)","gene_nuniq_Experiments"]].max(axis=1)
````

````python
print(df.loc[(df[["gene_nuniq_References (PMID)","gene_nuniq_Experiments"]].max(axis=1)>1)\
       |(~df["Support Type"].str.contains("weak",case=False))]["Target Gene (Entrez Gene ID)"].nunique())

df.loc[(df[["gene_nuniq_References (PMID)","gene_nuniq_Experiments"]].max(axis=1)>1)\
       |(~df["Support Type"].str.contains("weak",case=False)),"mir_target"] = 1
````

````python
# df.drop_duplicates(["Target Gene"])
````

````python
df_mir_gene = df.drop_duplicates(["Target Gene"])[["Target Gene","Target Gene (Entrez Gene ID)","mir_target"]]
df_mir_gene.to_csv("mirtar_human_gene_targetlist.csv",index=False)
````

#### additional data munge
* change ";" to make for easier tokenizing

````python
df = df.replace(";",". ",regex=True)
DROP_COLS = ["Target Gene (Entrez Gene ID)","Target Gene","Organism"]
df.drop(columns=DROP_COLS,errors="ignore",inplace=True)

df["total_nan"] = df.isna().sum(axis=1)
````
