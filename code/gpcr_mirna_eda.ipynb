```python
import pandas as pd
import seaborn as sns
import os
%matplotlib inline 
MIR_PATH = os.path.normpath(".")
```

```python
### from GPCR DB ,swissprot human, names: 
### Class B1 (Secretin):
b1_names = ["calcr","calrl","crfr1","crfr2","ghrhr","gipr","glp1r","glp2r","glr","sctr","pth1r","pth2r","pacr","vipr1","vipr2"]
```

```python
df = pd.read_csv(MIR_PATH+"/human_gene_mirtar_targets.csv.gz")#.set_index("Entry")
df["name"] = df['Entry name'].str.lower().str.split("_",expand=True)[0] # compatible with gpcr

df = df.filter(["name",'Protein names','Protein families',
       'mir_target', 'bool_mir_target', 'Length', 'Keywords',
       'Gene names  (primary )', 'Entry name', "Entry",
#        'Sequence', 'Tissue specificity'
               ],axis=1).drop_duplicates()

print(df.shape[0])
```

```python
df["Protein families"] = df["Protein families"].str.lower().str.split(" family,|superfamily,")
df = df.explode("Protein families").dropna(thresh=50,axis=1).drop_duplicates()
print(df.shape[0])
df
```

##### keep only receptors
8 +- gpcr

```python
df = df.loc[df["Protein families"].str.contains("GPCR|receptor",case=False,na=False)]
df
```

```python
df = df.loc[df["Protein families"].str.contains("GPCR|coupled receptor",case=False,na=False)].drop_duplicates("Protein names")
print(df.shape[0],"rows")
df["Protein families"].value_counts()
```
